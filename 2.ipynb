```python
import pandas as pd
```

```python
# Loading the petrol consumption dataset
data = pd.read_csv('petrol_consumption.csv')
```

```python
data2 = data.copy()
```

```python
# Splitting the data
X = data2.iloc[:,:-1]
Y = data2.iloc[:,4]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# 70:30 Split
X_train_70, X_test_70, Y_train_70, Y_test_70 = train_test_split(X, Y, random_state=42, test_size=0.3)
```

```python
# 80:20 Split
X_train_80, X_test_80, Y_train_80, Y_test_80 = train_test_split(X, Y, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Create the Multiple Linear Regression model for 70:30
mlr_model_70_30 = LinearRegression()
```

```python
# Create the Multiple Linear Regression model for 80:20
mlr_model_80_20 = LinearRegression()
```

```python
# Fit the model on the training data (X_train_70 and Y_train_70)
mlr_model_70_30.fit(X_train_70, Y_train_70)
```

```python
# Fit the model on the training data (X_train_80 and Y_train_80)
mlr_model_80_20.fit(X_train_80, Y_train_80)
```

```python
# predict the model using the 'X_test_70' dataset
Y_predictions_70 = mlr_model_70_30.predict(X_test_70)
```

```python
# predict the model using the 'X_test_80' dataset
Y_predictions_80 = mlr_model_80_20.predict(X_test_80)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
```

```python
# Evaluate the model performance (optional but important) 70:30
mse = mean_squared_error(Y_test_70, Y_predictions_70)
mae = mean_absolute_error(Y_test_70, Y_predictions_70)
r2 = r2_score(Y_test_70, Y_predictions_70)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```

```python
# Evaluate the model performance (optional but important) 80:20
mse = mean_squared_error(Y_test_80, Y_predictions_80)
mae = mean_absolute_error(Y_test_80, Y_predictions_80)
r2 = r2_score(Y_test_80, Y_predictions_80)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```

```python
# b) Results comparison
print("70:30 MAE:", mean_absolute_error(Y_test_70, Y_predictions_70))
print("80:20 MAE:", mean_absolute_error(Y_test_80, Y_predictions_80))
```
