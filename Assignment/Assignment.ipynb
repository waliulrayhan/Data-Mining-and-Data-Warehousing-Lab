```python
import pandas as pd
```

```python
# load the dataset 'heart.csv'
data = pd.read_csv('heart.csv')
```

```python
# show the first 5 rows of the dataset
data.head()
```

```python
# Heart Disease Dataset: Data Preprocessing
# Steps included: Label Encoding, One-Hot Encoding, Correlation Plot, and Standardization
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
# Apply Label Encoding (e.g., 'sex', 'thal')
label_encoder = LabelEncoder()
```

```python
data['sex'] = label_encoder.fit_transform(data['sex'])
data['thal'] = label_encoder.fit_transform(data['thal'])
```

```python
# Apply One-Hot Encoding (e.g., 'cp', 'slope', 'restecg')
data_encoded = pd.get_dummies(data, columns=['cp', 'slope', 'restecg'], drop_first=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Plot correlation heatmap
plt.figure(figsize=(12, 8))
correlation_matrix = data_encoded.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap of Heart Disease Dataset")
plt.show()
```

```python
# Apply Standardization to numerical features
scaler = StandardScaler()
```

```python
numerical_features = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'ca']
data_encoded[numerical_features] = scaler.fit_transform(data_encoded[numerical_features])
```

```python
# Save the processed dataset to a new CSV file
processed_file_path = 'heart_processed.csv'
data_encoded.to_csv(processed_file_path, index=False)
```
