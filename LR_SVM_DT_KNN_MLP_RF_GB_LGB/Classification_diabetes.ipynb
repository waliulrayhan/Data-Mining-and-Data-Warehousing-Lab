```python
# store all rows and all columns into X except the last column
X=data2.iloc[:,:-1]
```

```python
# store the last column into Y
Y=data2.iloc[:,8]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# split the data with 80:20 ratio as train set and test set
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, test_size=.2)
```

```python
# show the total number of rows and columns for 'X_train'
X_train.shape
```

```python
# show the total number of rows and columns for 'Y_train'
Y_train.shape
```

```python
# show the total number of rows and columns for 'Y_test'
Y_test.shape
```

```python
# show the 'X_train' data frame
X_train
```

```python
# show the 'Y_train' data frame
Y_train
```

```python
# import 'LogisticRegression' model
from sklearn.linear_model import LogisticRegression
```

```python
# create the 'LogisticRegression' model 
model=LogisticRegression(solver='lbfgs', max_iter=200)
```

```python
# fit 'X-train' and 'Y_train' into the model
model.fit(X_train, Y_train)
```

```python
# predict the model using the 'X_test' dataset
Y_predictions = model.predict(X_test)
```

```python
# show the 'Y_predictions' list
Y_predictions
```

```python
# import 'accuracy_score' and 'confusion_matrix'
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# calculate the accuracy of the model
accuracy_LR=accuracy_score(Y_test, Y_predictions)
```

```python
# show the accuracy score
accuracy_LR
```

```python
# calculate and show the confusion_matrix of the model
confusion_matrix(Y_test, Y_predictions)
```
